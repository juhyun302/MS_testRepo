```python
# -*- coding: utf-8 -*-
import pandas as pd
from pathlib import Path
import re

# === 경로 설정 ===
CSV_PATH = Path(r"C:\Users\EL19\OneDrive - (주)엘릭서\바탕 화면\음식\MS_testRepo\image_food_label_join.csv")  # 네가 받은 image_food_label_join.csv 경로
OUT_PATH = CSV_PATH.with_name("classes.txt")  # 같은 폴더에 classes.txt 생성

# === CSV 로드 ===
df = pd.read_csv(CSV_PATH)

# === '식품명(B열)'이 매칭된 것만 사용 (비어있거나 매칭 실패한 행은 제외) ===
mask = df["식품명(B열)"].notna() & (df["식품명(B열)"].astype(str).str.strip() != "")
names = df.loc[mask, "식품명(B열)"].astype(str).unique()

# === YOLO용 형식 맞추기: 최소 전처리 ===
cleaned = []
for n in names:
    x = n.strip()                # 앞뒤 공백 제거
    # 필요하면 공백/특수문자 제거해서 더 깔끔하게 쓸 수도 있음
    # x = re.sub(r"\s+", "", x)  # <- 공백 없애고 싶으면 주석 해제
    # x = re.sub(r"[\(\)※★!]", "", x)
    cleaned.append(x)

# === 최종 중복 제거 + 정렬 ===
classes = sorted(set(cleaned))

# === 저장 ===
with open(OUT_PATH, "w", encoding="utf-8") as f:
    f.write("\n".join(classes))

print("classes.txt 생성 완료:", OUT_PATH)
print("클래스 개수:", len(classes))
```
